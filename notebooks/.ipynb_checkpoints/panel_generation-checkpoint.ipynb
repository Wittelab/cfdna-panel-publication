```python
# Show plots within the noteboook
%matplotlib inline 

# Ignore warnings
import warnings
warnings.filterwarnings('ignore')

# Required system Libraries 
import sys, os, datetime, time
import random

# For plotting
import seaborn as sns
import matplotlib.pyplot as plt

# Numeric libraries
import pandas as pd
import numpy as np 

# orchid-ml
from orchid_ml import MutationMatrix
from orchid_ml import load_matrix


# Modeling options
class options:
    prostate_dbi = 'REDACTED'
    tag          = 'SVMv4_Enrichment_20171107' # A name for this dataset, available upon request
    use_saved    = True  # Used saved data (specified by tag) or generate new data (and tag)
    cutoff       = 275   # Number of donors to split on (275 is half for each class)
    n_features   = 20    # Desired number of features after selection
    dp_ratio     = None  # Donor to passenger ratio (use dataset ratio if None)
    model_limit  = 16000 # Max number of mutations to train with 
    cv           = 10    # Number of cross validation folds
    
# Generate a new dataset
if not options.use_saved:
    date = datetime.datetime.now().strftime("%Y%m%d")
    output_dir = "../"+options.tag+"_"+date
    try:
        os.stat(output_dir)
        print "WARNING: Will overwrite data in '%s' unless 'use_saved' is True" % output_dir
    except:
        os.mkdir(output_dir)
        print "Made job directory '%s'" % output_dir
# Or load an old one
else:
    output_dir = options.tag
    print "Using data from '%s'" % output_dir
    
os.chdir(output_dir)
```

```python
def make_colormap(seq):
    """Return a LinearSegmentedColormap
    seq: a sequence of floats and RGB-tuples. The floats should be increasing
    and in the interval (0,1).
    """
    seq = [(None,) * 3, 0.0] + list(seq) + [1.0, (None,) * 3]
    cdict = {'red': [], 'green': [], 'blue': []}
    for i, item in enumerate(seq):
        if isinstance(item, float):
            r1, g1, b1 = seq[i - 1]
            r2, g2, b2 = seq[i + 1]
            cdict['red'].append([item, r1, r2])
            cdict['green'].append([item, g1, g2])
            cdict['blue'].append([item, b1, b2])
    return mcolors.LinearSegmentedColormap('CustomMap', cdict)

## Some functions to help with color
# Convert an rgb value to a hex code
def to_hex(rgb):
    r,g,b = rgb
    return "#{0:02x}{1:02x}{2:02x}".format(int(255*r), int(255*g), int(255*b))
# Get x color codes spanning the colorspace evenly
def hexme(x=5):
    rgbs = sns.color_palette(options.palette, 100)[::int(100./(x-1))-1]
    return map(to_hex, rgbs)

def reverse_colormap(cmap, name = 'my_cmap_r'):      
    reverse = []
    k = []   
    for key in cmap._segmentdata:    
        k.append(key)
        channel = cmap._segmentdata[key]
        data = []
        for t in channel:                    
            data.append((1-t[0],t[2],t[1]))            
        reverse.append(sorted(data))    

    LinearL = dict(zip(k,reverse))
    my_cmap_r = mcolors.LinearSegmentedColormap(name, LinearL) 
    return my_cmap_r

cmap_r = reverse_colormap(cmap, name = 'my_cmap_r')
```

## Get donors to train

```python
# Get the number of mutations per donor (donor burdens)
donor_burdens = pd.read_sql(
    """
        SELECT donor_id, count(*) AS burden 
        FROM ssm 
        WHERE is_simulated=0
        GROUP BY donor_id 
        ORDER BY burden
    """,
    options.prostate_dbi
)
```

```python
# Get some additinal donor information
donor_info = pd.read_csv("donor_info.tsv", sep="\t")
#cols = ['icgc_donor_id', 'donor_tumour_stage_at_diagnosis_clean', 'donor_relapse_type_clean']
#donor_info = donor_info[cols]
#donor_info.columns = ['donor_id', 'stage', 'metastatic?']
donor_burdens = \
    donor_burdens.merge(donor_info, how='left', left_on='donor_id', right_on='icgc_donor_id')
```

```python
# Plot the burden distribution across patients
plt.figure(figsize=(16,12))
sns.set_context("poster")
sns.set(font_scale = 2)
sns.set_style("white")

hist_kws={'log':True, 'rwidth':0.75, 'edgecolor':'#333333', 'linewidth':2, 'alpha':0.50}
ax = sns.distplot(donor_burdens['burden'], 
                  hist_kws=hist_kws, kde=False, color="#eeeeee", bins=50)
ax.set(xlabel='Burden (# Mutations)', ylabel='Count')
sns.despine()
```

```python
# Get tumor burdens and divide into two equal sized groups, those with low and high burdens
#  Low burden is defined as patient tumors with less than 675 
#  High burden is defined as more than that
# This cutoff, corresponding to 10^2.83 mutations (675), is reasonable considering
#   the above figure suggests a cutoff of around 10^2.5 (316) would be appropriate

burdens = donor_burdens.sort_values(by='burden')
driver_enriched    = list(burdens.iloc[0:options.cutoff]['donor_id'])
passenger_enriched = list(burdens.iloc[-options.cutoff:len(burdens)+1]['donor_id'])

# Get the actual donor:passenger mutation ratio for training if one isn't specified
dp_ratio = options.dp_ratio
if not dp_ratio:
    dp_ratio = float(burdens.iloc[0:options.cutoff]['burden'].sum())\
               /burdens.iloc[-options.cutoff:len(burdens)+1]['burden'].sum()

# Print a summary
print "Driver enriched threshold:     < %5d mutations \n\
Passenger enriched threshold: >= %5d mutations" \
% (burdens['burden'].iloc[options.cutoff], burdens['burden'].iloc[-options.cutoff])
print "\nDriver enriched:    %3d donors totalling %6d mutations\n\
Passenger enriched: %3d donors totalling %6d mutations\n\
In both sets:       %3d donors " \
% (\
   len(driver_enriched), sum(burdens[burdens['donor_id'].isin(driver_enriched)]['burden']), \
   len(passenger_enriched), sum(burdens[burdens['donor_id'].isin(passenger_enriched)]['burden']), \
   len(set(driver_enriched).intersection(set(passenger_enriched)))
  )
print "Driver/Passenger ratio: %.3f\n" % dp_ratio
```

```python
# Classify patients based on burden and cutoff values
stage = 'donor_tumour_stage_at_diagnosis_clean'
burdens['log(burden)'] = np.log10(donor_burdens['burden'])
burdens['driver']=False
burdens.iloc[0:options.cutoff]['driver']=True
burdens[stage] = burdens[stage].str.strip("T").fillna(0).astype(int)
```

## Load or generate prostate donor data

```python
%%time
## Load previous data, or generate it with load() and encode()
if options.use_saved and os.path.exists("prostate23_mutations.pkl"):
    print "Loading prostate data..."
    mutations = load_matrix("prostate23_mutations.pkl")
else:
    print "Generating prostate data..."
    de = MutationMatrix(db_uri=options.prostate_dbi)
    de.load_mutations(by='donor', ids=driver_enriched)
    pe = MutationMatrix(db_uri=options.prostate_dbi)
    pe.load_mutations(by='donor', ids=passenger_enriched, number_real=de.shape[0])
    mutations = pd.concat([de,pe])
    mutations.load_features()
    mutations.encode(quiet=False)
    # Define 'is_enriched' as the mutation label to train with
    ## NOTE: In cases where mutations are in both sets, the will be labeled as drivers
    mutations['is_enriched'] = mutations['donor_id'].isin(driver_enriched)
    mutations.set_label_column('is_enriched')
    mutations.save_matrix("prostate23_mutations.pkl")
```

```python
mutations.groupby('is_enriched').size()
```

```python
# Show how many donors were loaded (this should be 550)
print "Donors at this point: %s" % len(set(mutations['donor_id']))
```

```python
# Get mutations from enriched and non-enriched donors
enriched = mutations[mutations['is_enriched']]
non      = mutations[~mutations['is_enriched']]

# Calculate roughly how many mutations to sample to reach the modeling limit
#  given the target d:p ratio
target_d    = int(dp_ratio*(options.model_limit/(dp_ratio+1)))
target_d    = min(len(set(enriched['mutation_id'])), target_d)
target_p    = int((1/dp_ratio)*target_d)

# And then downsample mutations from each to reach the model limit, maintaining the dp_ratio
enriched = enriched[enriched['mutation_id'].isin(list(enriched['mutation_id'].sample(target_d)))]
non      = non[non['mutation_id'].isin(list(non['mutation_id'].sample(target_p)))]
downsampled = pd.concat([enriched,non])
print "Number of downsampled mutations: %d" % downsampled.shape[0]
print "DE:PE ratio: %.4f" % dp_ratio
```

```python
%%time
# Now create a model using the full downsampled data (without cross validation)
if options.use_saved and os.path.exists("prostate23_model.pkl"):
    print "Loading prostate model..."
    full_model = load_matrix("prostate23_model.pkl")
else:
    print "Generating prostate model..."
    #full_model = MutationMatrix(db_uri=options.prostate_dbi)
    full_model = downsampled.copy()
    full_model.set_features(new_features)
    full_model.svm(cross_validate=False, test_size=0.0, class_weight=None)
    full_model.save_matrix("prostate23_model.pkl")
```

```python
%%time
# Apply the model to all mutations
if options.use_saved and os.path.exists("prostate23_final.pkl"):
    print "Loading prostate final scores..."
    all_mutations = load_matrix("prostate23_final.pkl")
else:
    print "Generating prostate final scores..."
    #all_mutations = MutationMatrix(db_uri=options.prostate_dbi)
    all_mutations = mutations.copy()
    # Use the same selected features
    all_mutations.set_features(list(full_model.features))
    # And normalize the same way
    X = full_model._normalize(all_mutations[full_model.features], 
                              imputer=full_model.imputer, 
                              scaler=full_model.scaler, 
                              as_df=True)[0]
    # Determine distances from the discriminatory hyperplane
    all_mutations['Distance'] = full_model.model.decision_function(X)
    all_mutations = all_mutations.sort_values(by='Distance')    
    all_mutations.save_matrix('prostate23_final.pkl')
```
