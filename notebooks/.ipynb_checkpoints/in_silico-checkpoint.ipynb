```python
%matplotlib inline  
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib.pyplot as plt
from matplotlib.offsetbox import AnchoredText

from slurpie import *

import warnings
warnings.filterwarnings("ignore")
```

## Load

### Get Tumor/Normal variants

##### SpeedSeq T/N Panel data (uses freebayes to call variants)

```python
## Run using the tn-SpeedSeq nextflow script
## Filters applied by the SpeedSeq workflow (see freebayes documentation https://github.com/ekg/freebayes)
## freebayes -f /wittelab/data1/hg19/ucsc.hg19.fasta 
##     --pooled-discrete 
##     --genotype-qualities 
##     --min-repeat-entropy 1 
##     --min-alternate-fraction 0.05 
##     --min-alternate-count 2
##     normal.bam
##     tumor.bam
## Additionally, use only variants flagged as PASS by freebayes
dir_ = "../TN/panel-speed/*.vcf.gz"
tumor_variants = read_freebayes_tn(dir_)
tumor_unfiltered = tumor_variants.copy()
tumor_variants   = tumor_unfiltered[(tumor_unfiltered['Focus Filter']=='PASS')]
```

### Get cfDNA variants  

```python
## Filters applied during Curio analysis
## Quality Threshold: Phred 30 (99.9%) kept
## UMI/UMT Family Consensus Threshold: 75% of reads in family must match
## UMI/UMT Minimum Family Size: Exclude families with less than 4 reads
## Minimum Coverage: Exclude loci with less than 4 families total
## Minimum Rare Allele Frequency: At least one family at the locus should show the rare allele
## Variants over 20% are not considered (possibly hets)
## Hamming distance of 2

## Additional filters imposed by the read_curio function in slurpie.py
## min_fam -- At least 2 families at the locus should show the rare allele (stricter than Curio)
 
# Load the cfDNA variants
dir_="../cfdna/curio/*.csv"
curio = read_curio(dir=dir_, filter_=True, min_fam=2, rare_only=True)
#unfiltered_curio = curio.copy()
#curio = filter_variants(curio)
```

```python
## If multiple cfDNA variant sets were loaded above, use this one from here on out
cfDNA_variants = curio
```

```python
# Get list of patients in either dataset
cfDNA_pts = set(cfDNA_variants['Patient'])
tumor_pts = set(tumor_variants['Patient'])
both_pts  = cfDNA_pts.intersection(tumor_pts)
```

```python
## A list of patients that have both T/N and cfDNA data
both_pts
```

### Mutations Per Foci

```python
# Get the number of reads for each patient (from Curio)
file_ = "../cfDNA/sample_reads.tsv"
num_reads = read_num_reads(file=file_)
```

**Variants in multiple foci**

```python
# How to define a unique variant
ucd = ['Patient','Chromosome','Position','Reference','Variant']
```

```python
# Get the number of foci per patient
num_pt_foci = pd.DataFrame(
    tumor_variants[['Patient','Focus']]\
        .drop_duplicates()\
        .groupby(['Patient'])\
        .size(), columns=['Count'])
```

```python
# Get the number of foci in which each variant is present and calculate percent clonality 
foci_presence = pd.DataFrame(
    tumor_variants[['Focus']+ucd].drop_duplicates()\
        .groupby(ucd, as_index=False)\
        .size(),
    columns=['Num Foci Present']).reset_index()
# Add the total number of foci per patient as a column and calculate foci presence %s
foci_presence = foci_presence.merge(num_pt_foci, how="left", left_on='Patient', right_index=True)
foci_presence['Percent'] = foci_presence['Num Foci Present']/foci_presence['Count']
## Add back the rest of the variant info:
##   (Focus, ID, Quality, Focus Filter, Focus Information, Format, Normal, Tumor)
foci_presence = tumor_variants.merge(foci_presence, on=ucd)
```

```python
## Get the number of unique mutations found in more than one foci per patient 
tn_can = foci_presence.drop_duplicates(ucd)
tn_can = tn_can[tn_can['Num Foci Present']>1]
tn_can = tn_can.groupby('Patient').size()
tn_can = pd.DataFrame(tn_can, columns=['Total'])
```

```python
detected = cfDNA_variants.merge(foci_presence, how='inner', left_on=ucd, right_on=ucd)
```

```python
meta = ['Focus','Num Foci Present','Count','Percent','Frequency']
# Get a list of mutations detected in BOTH cfDNA and these (Tumor-Normal) tumor_variants
detected = cfDNA_variants.merge(foci_presence, how='inner', left_on=ucd, right_on=ucd)
detected = detected.merge(num_reads, how='left', left_on='Patient', right_index=True)
detected = detected.drop_duplicates(ucd)[ucd+meta]
detected['Frequency'] = detected['Frequency']/100
detected['Log Frequency'] = np.log10(detected['Frequency'])
```

```python
gt1_foci_counts ## Used for figure 4A
```

```python
## For each patient, count the number of variants in one or more foci
tt = pd.DataFrame(detected.groupby(ucd).size(), columns=['Count'])
one_foci_counts = pd.DataFrame(tt.reset_index().groupby('Patient').size(), columns=["one"])

## and greater than one focus
gt1_foci = detected[detected['Num Foci Present']>1]
uu = pd.DataFrame(gt1_foci.groupby(ucd).size(), columns=['Count'])
gt1_foci_counts = pd.DataFrame(uu.reset_index().groupby('Patient').size(), columns=["gt1"])

## and in all foci
all_foci = detected[detected['Percent']==1.0]
vv = pd.DataFrame(all_foci.groupby(ucd).size(), columns=['Count'])
all_foci_counts = pd.DataFrame(vv.reset_index().groupby('Patient').size(), columns=["all"])

## Merged dataset
fcs = one_foci_counts.merge(gt1_foci_counts, how="outer", right_index=True, left_index=True)\
                     .merge(all_foci_counts, how="outer", right_index=True, left_index=True)\
                     .fillna(0)
```

```python
stacked = pd.DataFrame(fcs.stack()).reset_index()
stacked.columns = ['Patient', 'Dataset', 'Count']

sns.set_context("poster")
sns.set_style("white")
fig, ax = plt.subplots(figsize=(11.7, 8.27))

sns.barplot(x='Patient', y='Count', hue='Dataset', data = stacked, linewidth=5, palette=[pal[2],pal[4],pal[7]])

plt.xticks(rotation=45, fontsize = 15);
sns.despine()

ax.set_title("Foci Shared Somatic cfDNA Detected Variants")
plt.show()
```

```python
from matplotlib.patches import Rectangle

sns.set_context("poster")
sns.set_style("white")
fig, ax = plt.subplots(figsize=(11.7, 8.27))


# Show cfDNA variants with no tumor support
#sns.swarmplot(x='Patient', y='Log Frequency', data=cfDNA_only, 
#              size=3, linewidth=0, color="#2e3436", alpha=.35)
# Show cfDNA variants detected that are not in T/N 
#sns.swarmplot(x='Patient', y='Log Frequency', data=reduced[~reduced['Detected']], 
#              size=3, linewidth=0, color="#2e3436", alpha=.35)
# Show detected in both
sns.swarmplot(x='Patient', y='Log Frequency', data=detected, 
              size=5, linewidth=0, alpha=.65, palette=pal) #color=flatui[7]
ax=sns.boxplot(x='Patient', y='Log Frequency', data=detected, 
                whis=1.5, linewidth=2, fliersize=0, palette=pal)
plt.xticks(rotation=45, fontsize = 15);

# Add germline patch
#                       x1          max AF  width         height
ax.add_patch(Rectangle((-1, np.log10(1.1)), len(set(detected['Patient']))+1, np.log10(.30), facecolor="#eeeeee", alpha=1))
# Theoretical WGS limit = 32.5%
plt.axhline(y=np.log10(0.325),  xmin=0, xmax=1, hold=None, 
            linestyle='dashed', linewidth=1, color=flatui[2])
# Theoretical WES limit = 2.75%
#plt.axhline(y=np.log10(0.0275), xmin=0, xmax=1, hold=None, 
#            linestyle='dashed', linewidth=1, color=flatui[0])
# Theoretical Panel limit = 0.8% with 2,500X @ 10reads/family, 2 family coverage
plt.axhline(y=np.log10(0.008),  xmin=0, xmax=1, hold=None, 
            linestyle='dashed', linewidth=1, color=flatui[7])


for i,artist in enumerate(ax.artists):
    # Set the linecolor on the artist to the facecolor, and set the facecolor to None
    col = artist.get_facecolor()
    artist.set_edgecolor(col)
    artist.set_facecolor('None')
    artist.set_zorder(10)

    # Each box has 6 associated Line2D objects (to make the whiskers, fliers, etc.)
    # Loop over them here, and use the same colour as above
    for j in range(i*6,i*6+6):
        line = ax.lines[j]
        line.set_color(col)
        line.set_mfc(col)
        line.set_mec(col)

ax.set(xlabel='Patient', ylabel='Allele Frequency (log-scaled)')
plt.yticks(
    [0, -1.000, -2.000, -3.000], 
    [1,  0.100,  0.010,  0.001]
)
sns.despine()
plt.show()
```
